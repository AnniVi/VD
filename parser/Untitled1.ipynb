```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from flask import Flask, render_template, request
import io
import base64
```

```python
data = pd.read_csv(filepath_or_buffer='./dataframe.csv',
                    sep=',', encoding='UTF-8')
```

```python
data = data.dropna(axis=0, how='any', subset=['brand'])

data = data.loc[data['brand'] != 'Сменная']

data = data.loc[data['brand'] != 'Планшет']

data = data.loc[data['brand'] != 'Мобильный']

data.loc[data['brand']=='Lenovoen', 'brand'] = 'Lenovo'

data['discounted_goods'] = data['discounted_goods'].astype(bool)

data['dual_sim'] = data['dual_sim'].astype(bool)
```

```python
data.loc[data['brand'] == 'Tp-Link']
```

```python
data1 = data.copy()
data1 = data1.dropna(subset=['brand', 'name', 'cost'])
data1.shape
```

```python
df = data1.groupby(['memory_size_Gb', 'brand'])['cost'].mean()

df1 = pd.DataFrame({'price_mean': df}).reset_index()

df2 = pd.pivot_table(df1, index='memory_size_Gb', columns='brand', values='price_mean')
```

```python
df = data1.groupby(['color', 'year'])['cost'].mean()

df1 = pd.DataFrame({'price_mean': df}).reset_index()

df2 = pd.pivot_table(df1, index='year', columns='color', values='price_mean')
```

```python
app = Flask(__name__)

@app.route("/", methods=['POST', 'GET'])
def hello():
    if request.method == 'POST':
        if request.form['submit'] == 'Heat_map':
            plt.clf()
            plt.figure(figsize=(12,12))
            sns.heatmap(df2)
            output = io.BytesIO()
            plt.savefig(output, format='png')
            plot_url = base64.b64encode(output.getvalue()).decode()
            output.close()
            return render_template('index.html', graphic=plot_url)
        elif request.form['submit'] == 'Bar':
            df2.plot(kind='bar', rot=45)
            df2.plot(figsize=(12,12));
            output = io.BytesIO()
            plt.savefig(output, format='png')
            plot_url = base64.b64encode(output.getvalue()).decode()
            output.close()
        else:
            return "error"
    elif request.method == 'GET':
        plt.figure(figsize=(12,12))
        sns.heatmap(df2)
        output = io.BytesIO()
        plt.savefig(output, format='png')
        plot_url = base64.b64encode(output.getvalue()).decode()
        output.close()
    return  render_template('index.html', graphic = plot_url)

if __name__ == '__main__':
    app.run()
```
